```python
import glob
import cv2
import random as random
import matplotlib.pyplot as plt
import numpy as np

```

```python
base_path = "/content/drive/My Drive/TACAI/"
#take the path for all the jpg 
image_path_by_class =[]

for i in range(4):
  c_p = base_path + str(i) + "/*.jpg"
  image_path_by_class.append(glob.glob(c_p))

print(image_path_by_class)
```

```python
# shuffle all the images to be random 
for a in image_path_by_class:
  random.shuffle(a, lambda : .5)
print(image_path_by_class)
```

```python
# impartire date in train si test 70% 30%

train_lb = []
test_lb = []
for a in image_path_by_class:
  cls_t =[]
  cls_v = []
  for i,e in enumerate(a):
    if i< 7 :
      cls_t.append(e)
    else :
      cls_v.append(e)
  train_lb.append(cls_t)
  test_lb.append(cls_v)

print(train_lb)
print(test_lb)
```

```python
# incarcare imagini

train = []
test =[]

for cls in train_lb :
  tmp = []

  for e in cls :
    im = cv2.imread(e)
    tmp.append(im)
  train.append(tmp) 

for cls in test_lb :
  tmp = []

  for e in cls :
    im = cv2.imread(e)
    tmp.append(im)
  test.append(tmp) 
```

```python
# afisare imagini

plt.figure(), plt.imshow(train[0][0][...,::-1]) # ptc cv2 citeste BGR si imshow printeaza RGB
plt.figure(), plt.imshow(test[0][0][...,::-1])
```

```python
print(np.shape(test[0][0])) # imagini sunt mari pentru a lucra in timp 
```

```python
for i,cls in enumerate(train) :
  for j, im in enumerate(cls) :
    r = cv2.resize(im,(480,640))
    train[i][j] = r

for i,cls in enumerate(test) :
  for j, im in enumerate(cls) :
    r = cv2.resize(im,(480,640))
    test[i][j] = r
    
plt.figure(), plt.imshow(train[0][0][...,::-1]) # ptc cv2 citeste BGR si imshow printeaza RGB
plt.figure(), plt.imshow(test[0][0][...,::-1])
```

```python
# realizare histograma

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([train[0][0][...,::-1]],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
train_g = []
test_g = []

for cls in train:
  tmp = []
  for im in cls :
    mean = np.mean((im[:,:,1],im[:,:,2]), axis =0).astype(np.uint8)
    tmp.append(mean)
  train_g.append(tmp)

for cls in test:
  tmp = []
  for im in cls :
    mean = np.mean((im[:,:,1],im[:,:,2]), axis =0).astype(np.uint8)
    tmp.append(mean)
  test_g.append(tmp)

plt.figure(), plt.imshow(train_g[0][0], cmap = 'gray')
plt.figure(), plt.imshow(test_g[0][0], cmap ='gray')
```
